```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
video_list = pd.read_csv('data/video_list.csv')
video_list
```

```python
df = pd.read_json('data/DEAP_features_updated.json')
df.head()
```

```python
VAQs = []
VAQs_class = []
for index, row in df.iterrows():
    video = video_list[video_list.Experiment_id == row.session]
    #print(video.VAQ_Estimate)
    #print(index, row['subject_id'], row['session'])
    q_n = video['VAQ_Estimate'].values[0]
    VAQs.append(q_n)
    if q_n == 1:
        VAQs_class.append('HAHV')
    elif q_n == 2:
        VAQs_class.append('LAHV')
    elif q_n == 3:
        VAQs_class.append('LALV')
    elif q_n == 4:
        VAQs_class.append('HALV')
df['VAQ_Estimate'] = VAQs
df['VAQ_Estimate_class'] = VAQs_class
df.head()
```

### Why I think Alex stinks

1. He looks at my screen

```python
a_L = []
v_L = []
va_L = []
threshold = 5
for index, row in df.iterrows():
    valence = row["valence"]
    arousal = row["arousal"]
    if valence >= threshold:
        v_L.append("HV")
    elif valence < threshold:
        v_L.append("LV")
        
    if arousal >= threshold:
        a_L.append("HA")
    elif arousal < threshold:
        a_L.append("LA")
        
    va_L.append(a_L[index] + v_L[index])

df['a_labels'] = a_L
df['v_labels'] = v_L
df['va_labels'] = va_L
df.head()
```

```python
df.to_json('data/DEAP_features_updated.json')
```

```python
plt.rcParams["figure.figsize"] = [20, 4]

VAQ = 'va_labels'
# Boxplot distribution of the participants subjective ratings per scale
lalv = df[df[VAQ] == 'LALV']
halv = df[df[VAQ] == 'HALV']
lahv = df[df[VAQ] == 'LAHV']
hahv = df[df[VAQ] == 'HAHV']

fig, axs = plt.subplots(1, 4)

# plot each column in a separate subplot
axs[0].boxplot([lalv['liking'], lalv['valence'], lalv['arousal'], lalv['dominance']], labels=['L', 'V', 'A', 'D'])
axs[0].set_xlabel('LALV')
axs[0].set_ylabel('Self assessment')
axs[0].axhline(y=5, color='r', linestyle='--')  # add horizontal line at y=5
axs[1].boxplot([halv['liking'], halv['valence'], halv['arousal'], halv['dominance']], labels=['L', 'V', 'A', 'D'])
axs[1].set_xlabel('HALV')
axs[1].axhline(y=5, color='r', linestyle='--')  # add horizontal line at y=5
axs[2].boxplot([lahv['liking'], lahv['valence'], lahv['arousal'], lahv['dominance']], labels=['L', 'V', 'A', 'D'])
axs[2].set_xlabel('LAHV')
axs[2].axhline(y=5, color='r', linestyle='--')  # add horizontal line at y=5
axs[3].boxplot([hahv['liking'], hahv['valence'], hahv['arousal'], hahv['dominance']], labels=['L', 'V', 'A', 'D'])
axs[3].set_xlabel('HAHV')
axs[3].axhline(y=5, color='r', linestyle='--')  # add horizontal line at y=5


# add a title to the entire figure
fig.suptitle('Rating distributions for the emotion induction conditions')
fig.text(0.5, 0.01, 'Scales by condition', ha='center')
plt.show()
```
