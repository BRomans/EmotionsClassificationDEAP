```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [10, 10]


def plot_vadl_quadrant(df, subject_id, scale_factor=20):
    df_subject = df[df.subject_id == subject_id].sort_values('session').reset_index(drop=True)
    df_labels = df_subject[df_subject.record_type == 'Acquisition'].reset_index(drop=True)
    hahv = []
    lahv = []
    halv = []
    lalv = []
    valence = df_labels.valence
    arousal = df_labels.arousal
    dominance = df_labels.dominance
    liking = df_labels.liking

    for i in range(0, len(df_labels.session)):
        if valence[i] >= 5 and arousal[i] >= 5:
            hahv.append([valence[i], arousal[i], dominance[i] * scale_factor, liking[i]])
        elif valence[i] >= 5 and arousal[i] < 5:
            lahv.append([valence[i], arousal[i], dominance[i] * scale_factor, liking[i]])
        elif valence[i] < 5 and arousal[i] >= 5:
            halv.append([valence[i], arousal[i], dominance[i] * scale_factor, liking[i]])
        elif valence[i] < 5 and arousal[i] < 5:
            lalv.append([valence[i], arousal[i], dominance[i] * scale_factor, liking[i]])

    hahv = np.array(hahv)
    lahv = np.array(lahv)
    halv = np.array(halv)
    lalv = np.array(lalv)

    #havv = hahv - np.median(hahv, axis=1, keepdims=True)
    #lahv = lahv - np.median(lahv, axis=1, keepdims=True)
    #halv = halv - np.median(halv, axis=1, keepdims=True)
    #lalv = lalv - np.median(lalv, axis=1, keepdims=True)
    
    if len(hahv) > 0:
        plt.scatter(hahv[:,0], hahv[:,1], marker='^', sizes=hahv[:,2], c=hahv[:,3], label='HAHV', cmap='Wistia')
    if len(lahv) > 0:
        plt.scatter(lahv[:,0], lahv[:,1], marker='>', sizes=lahv[:,2], c=lahv[:,3], label='LAHV', cmap='Greens')
    if len(halv) > 0:
        plt.scatter(halv[:,0], halv[:,1], marker='<', sizes=halv[:,2], c=halv[:,3], label='HALV', cmap='Reds')
    if len(lalv) > 0:
        plt.scatter(lalv[:,0], lalv[:,1], marker='v', sizes=lalv[:,2], c=lalv[:,3], label='LALV', cmap='Blues')

    plt.ylabel("Arousal")
    plt.xlabel("Valence")
    plt.grid(True, linestyle='--', linewidth=0.5, color='gray')
    plt.xlim([0, 10])
    plt.ylim([0, 10])
    plt.legend()
    plt.title("Valence-Arousal-Dominance ratings for subject " + subject_id)
    plt.show()
    
#for sid in df.subject_id.unique():
sid = 's32'
plot_vadl_quadrant(df, subject_id=sid)
```

```python
from sklearn.preprocessing import MinMaxScaler


subject_ids = np.array(df.subject_id.unique())
subject_ids.sort()
new_df = pd.DataFrame()

for subject_id in subject_ids:
    subject_df = df[df.subject_id == subject_id]
    valence_ratings = np.array(subject_df.valence)
    og_shape = valence_ratings.shape
    scaler = MinMaxScaler(feature_range=(-1, 1))
    scaled_valence_ratings = scaler.fit_transform(valence_ratings.reshape(-1, 1))
    scaled_valence_ratings = scaled_valence_ratings.reshape(og_shape)
    subject_df = subject_df.assign(scaled_valence=scaled_valence_ratings)
    
    arousal_ratings = np.array(subject_df.arousal)
    og_shape = arousal_ratings.shape
    scaler = MinMaxScaler(feature_range=(-1, 1))
    scaled_arousal_ratings = scaler.fit_transform(arousal_ratings.reshape(-1, 1))
    scaled_arousal_ratings = scaled_arousal_ratings.reshape(og_shape)
    subject_df = subject_df.assign(scaled_arousal=scaled_arousal_ratings)
    
    new_df = pd.concat([new_df, subject_df], axis=0)
```

```python
new_df.to_json('data/DEAP_features_updated.json', orient='index')
```
